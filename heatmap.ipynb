```python

import pandas as pd
import geopandas as gpd
from shapely import geos
from shapely.geometry import Point
import fiona
import matplotlib.pyplot as plt
from fiona.crs import from_epsg,from_string
```

```python
data = pd.read_csv('../data/' + 'MissingMigrants-Global-2019-10-31T03-11-36.csv')
```

```python
data['lat']=data['Location Coordinates'].astype('str').apply(lambda x: x.split(',')).map(lambda x: x[0]).astype('float64') 
data['long']=data['Location Coordinates'].astype('str').apply(lambda x: x.split(',')).map(lambda x: x[-1]).astype('float64')
```

```python
data2 = data[['Location Coordinates','Region of Incident','Reported Year','Total Dead and Missing','lat','long']]
data2geometry = [Point(xy) for xy in zip(data2['long'], data2['lat'])]
crs = {'init': 'epsg:4326'}
gdf = gpd.GeoDataFrame(data2, crs=crs, geometry=data2geometry)
```

```python
import folium #, jinja2, vincent
from folium import plugins
from IPython.display import IFrame
from IPython.core.display import HTML
print(folium.__file__)
print(folium.__version__)
```

```python
#heatmap
nodes = [(gdf.lat[i], gdf.long[i]) for i in gdf.index]
map_osm = folium.Map(location=[gdf.lat.mean(), gdf.long.mean() ], tiles = 'cartodbpositron',
                     #tiles='stamentoner',
                     zoom_start=3,radius=8, max_zoom=13)
map_osm.add_child(plugins.HeatMap(nodes))
map_osm.fit_bounds(map_osm.get_bounds())
map_osm
```
